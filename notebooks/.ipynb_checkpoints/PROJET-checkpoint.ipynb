# **Phase de TRANSFORMATION et PREPARATION des données**

#1 - Format

Grâce à l'EDA, on sait que TotalCharges est sous le mauvais format, on va essayer de comprendre et arranger cela.

Si ça se trouve c'est purement une erreur de format :

````python
df.TotalCharges.astype(float)
````

On comprend que c'est fait exprès : certaines valeurs sont rentrées avec le symbole de monnaie, identifions ces cas :

````python
#On convertit les éléments récalcitrants en Nan : on obtients un df avec les vrais Nan de base ET les récalcitrants
df_nan_et_recalcitrants = pd.to_numeric(df['TotalCharges'], errors= 'coerce')

#Les éléments récalcitrants sont ceux qui n'était pas Nan au début et qui le sont devenus après conversion
recalcitrants = df[(df.TotalCharges.notna()) & (df_nan_et_recalcitrants.isna())]

recalcitrants.TotalCharges.sample(20)
````

### La toutes les valeurs semblent en dollars, ça va. SI ce n'était pas le cas :



*   Soit je supprime ces 3400 lignes (5% du jeu de données) -> va forcément impacter le modèle, même légèrement
*   Soit je supprime les caractères non numérique puis je convertis ; problème : il se peut que j'ai "30 dinars" et "30 euros", après transformation j'aurais 30 et 30, et le modèle lui ne fait pas de distinction et va donner le même poids à ces deux valeurs.

*   Une solution peut être de convertir moi même chacune des valeurs sous une même unité via un convertisseur monétaire






````python
# il y a beaucoup de caractères unique

# D'abord, on remplace d'abord la virgule par un point  (sinon on aura des valeurs extrêmes)
df.loc[recalcitrants.index, 'TotalCharges'] = recalcitrants['TotalCharges'].str.replace(',', '.', regex=False)

#On remplace tout ce qui n'est pas un chiffre ou un point par "rien"
# [^0-9.] signifie : "tout sauf les chiffres de 0 à 9 et le point"
df.loc[recalcitrants.index, 'TotalCharges'] = recalcitrants['TotalCharges'].str.replace(r'[^0-9.]', '', regex=True)

#on peut maintenant convertir en numérique sans avoir à transformer en NaN des valeurs qui donne de l'information
df.TotalCharges = pd.to_numeric(df['TotalCharges'], errors='coerce')

df.TotalCharges
````

````python
# --- définition des colonnes (à exclure : customerID, Churn = cible) ---
col_tenure   = ['tenure']
col_robust   = ['MonthlyCharges', 'TotalCharges']
col_cat      = X_train.select_dtypes(include='object').drop(columns='customerID').columns.tolist()

# --- pipelines par groupe de colonnes ---
pipe_tenure = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

pipe_robust = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())
])

pipe_cat = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# --- assemblage : chaque groupe de colonnes reçoit son propre traitement ---
preprocessor = ColumnTransformer([
    ('tenure', pipe_tenure, col_tenure),
    ('robust', pipe_robust, col_robust),
    ('cat',    pipe_cat,    col_cat)
], remainder='drop')  # drop customerID explicitement

# --- fit UNIQUEMENT sur train, transform sur train + test ---
X_train = preprocessor.fit_transform(X_train)
X_test  = preprocessor.transform(X_test)

# pour retrouver les noms de colonnes après transformation (utile pour coef_ ou feature_importances_)
feature_names = preprocessor.get_feature_names_out()
````

# **Modélisation**





# Objectif
 
L'objectif n'est pas de maximiser l'accuracy ou le F1-score, mais de **minimiser le coût moyen des erreurs de classification**.
 
On considère la matrice de perte suivante :
 
$$
L(y,\hat{y}) =
\begin{pmatrix}
0 & 25\\
200 & 0
\end{pmatrix}
$$
 
où :
 
- une fausse alerte (False Positive) coûte **25 €** ;
- un faux négatif (False Negative) coûte **200 €**.
Le risque associé à un classifieur $g$ est alors
 
$$
R(g) = \mathbb{E}\left[L(Y,g(X))\right].
$$
 
Cette quantité représente **le coût moyen attendu par individu**.
 
Notre objectif est donc de construire le classifieur
 
$$
g^{\star} = \underset{g}{\operatorname{argmin}} \; R(g).
$$
 
En pratique, l'espérance étant inconnue, on l'estime sur le jeu de test par
 
$$
\widehat{R}(g) = \frac{1}{n} \sum_{i=1}^{n} L\!\left(y_i,g(x_i)\right),
$$
 
qui correspond au **coût moyen empirique** (en euros).
 
---
 
# Pourquoi ne peut-on pas minimiser directement cette perte ?
 
La perte précédente s'écrit
 
$$
L(y,g(x)) = 25\,\mathbf{1}_{\{y=0,\;g(x)=1\}} + 200\,\mathbf{1}_{\{y=1,\;g(x)=0\}}.
$$
 
Elle dépend d'une fonction indicatrice.
 
Cette fonction est :
 
- discontinue ;
- non dérivable ;
- non convexe.
Les algorithmes d'optimisation utilisés en apprentissage automatique (gradient, Newton, LBFGS...) nécessitent une fonction différentiable.
 
Il est donc impossible de minimiser directement cette perte.
 
---
 
# Principe des pertes de substitution
 
Pour contourner ce problème, on remplace la perte précédente par une **fonction de substitution** (*surrogate loss*).
 
Dans le cas de la régression logistique, cette fonction est la **log-loss**
 
$$
\ell(y,p) = -\left[y\log(p) + (1-y)\log(1-p)\right],
$$
 
où
 
$$
p = P(Y=1\mid X=x).
$$
 
La log-loss est :
 
- convexe ;
- différentiable ;
- optimisable efficacement.
La régression logistique apprend donc ses paramètres en minimisant cette fonction plutôt que la perte de coût.
 
---
 
# Intégration des coûts dans l'apprentissage
 
Comme une erreur sur la classe 1 coûte huit fois plus cher qu'une erreur sur la classe 0, on pondère la log-loss.
 
On choisit
 
$$
w_0 = 25, \qquad w_1 = 200.
$$
 
La fonction effectivement minimisée devient
 
$$
\mathcal{L}(\theta) = \sum_{i=1}^{n} w_{y_i} \; \ell\!\left(y_i,p_i\right),
$$
 
avec
 
$$
p_i = P(Y=1\mid X=x_i).
$$
 
En Python, cela correspond à
 
```python
LogisticRegression(class_weight={0: 25, 1: 200})
```
 
Le solveur ne minimise donc **pas** directement le coût en euros.
 
Il minimise une approximation convexe de ce coût, beaucoup plus facile à optimiser.
 
---
 
# Théorie de la décision de Bayes
 
Une fois les probabilités
 
$$
\eta(x) = P(Y=1\mid X=x)
$$
 
estimées, le classifieur minimisant l'espérance de la perte est donné par la règle de Bayes.
 
On prédit la classe 1 lorsque
 
$$
25(1-\eta(x)) < 200\,\eta(x).
$$
 
Après simplification,
 
$$
\eta(x) > \frac{25}{25+200} = \frac{1}{9}.
$$
 
Le classifieur optimal est donc
 
$$
g^{\star}(x) =
\begin{cases}
1 & \text{si } \eta(x) > \dfrac{1}{9},\\
0 & \text{sinon.}
\end{cases}
$$
 
Le seuil optimal n'est donc plus $0.5$, mais
 
$$
\boxed{\dfrac{1}{9}}.
$$
 
---
 
# Évaluation des modèles
 
Il est important de distinguer **l'objectif d'optimisation** et **le critère d'évaluation**.
 
Pendant l'entraînement, la régression logistique minimise une log-loss (classique ou pondérée).
 
En revanche, les modèles seront comparés selon le véritable critère d'intérêt :
 
$$
\widehat{R}(g) = \frac{1}{n} \sum_{i=1}^{n} L\!\left(y_i,g(x_i)\right).
$$
 
Cette quantité représente le coût moyen observé sur le jeu de test.
 
C'est cette valeur qui permet d'évaluer les performances économiques d'un modèle.
 
---
 
# Plan expérimental
 
Trois approches seront comparées.
 
## 1. Régression logistique classique
 
Le modèle minimise la log-loss classique.
 
La décision est obtenue avec le seuil par défaut :
 
$$
\hat{y} = \mathbf{1}\left(\hat{\eta}(x) > 0.5\right).
$$
 
---
 
## 2. Régression logistique classique + seuil de Bayes
 
Le modèle est exactement le même.
 
Seule la règle de décision est modifiée :
 
$$
\hat{y} = \mathbf{1}\left(\hat{\eta}(x) > \frac{1}{9}\right).
$$
 
Cette approche applique directement la règle théorique minimisant l'espérance de la perte.
 
---
 
## 3. Régression logistique pondérée
 
Le modèle minimise une log-loss pondérée grâce à
 
```python
class_weight={0: 25, 1: 200}
```
 
Les coefficients appris sont donc différents.
 
La décision est ensuite prise avec le seuil par défaut $0.5$.
 
---
 
# Comparaison finale
 
Les trois modèles seront comparés selon un unique indicateur :
 
$$
\widehat{R}(g) = \frac{1}{n} \sum_{i=1}^{n} L\!\left(y_i,g(x_i)\right).
$$
 
Ce critère estime directement **le coût moyen en euros** généré par chaque stratégie de classification.
 
Le meilleur modèle sera donc celui qui minimise cette estimation empirique de
 
$$
\mathbb{E}\left[L(Y,g(X))\right].
$$


### Tout d'abord on va créer un estimateur de l'espérance de ma métrique, pour estimer ma perte moyenne en euro.

C'est un estimateur sans biais

````python
def weighted_loss(y_true, y_pred):
    """
    Renvoie une estimation de la perte moyenne en euro, pour un modèle donné
    """
    
    return np.mean(
        25 * ((y_true == 0) & (y_pred == 1))
        + 200 * ((y_true == 1) & (y_pred == 0))
    )
````

````python
"""
Comparaison de 3 stratégies de classification avec coûts asymétriques :
- False Positive (FP) : 25 €
- False Negative (FN) : 200 €

Variables attendues en entrée : X_train, X_test, Y_train, Y_test
(déjà encodées et standardisées)
"""



# ---------------------------------------------------------
# Coûts de la matrice de perte
# ---------------------------------------------------------
COST_FP = 25
COST_FN = 200
SEUIL_BAYES = COST_FP / (COST_FP + COST_FN)  # = 1/9

print(f"Seuil de Bayes optimal théorique : {SEUIL_BAYES:.4f}")


# ---------------------------------------------------------
# Fonction de coût moyen empirique R_hat(g)
# ---------------------------------------------------------
def cout_moyen(y_true, y_pred, cout_fp=COST_FP, cout_fn=COST_FN):
    """
    Calcule le coût moyen empirique R_hat(g) en euros.
    Retourne aussi la matrice de confusion pour le détail.
    """
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    cout_total = fp * cout_fp + fn * cout_fn
    n = len(y_true)
    cout_moyen_par_individu = cout_total / n

    return {
        "cout_total": cout_total,
        "cout_moyen": cout_moyen_par_individu,
        "TN": tn, "FP": fp, "FN": fn, "TP": tp,
        "n": n,
    }


def afficher_resultats(nom, resultats):
    print(f"\n--- {nom} ---")
    print(f"  TN={resultats['TN']}  FP={resultats['FP']}  "
          f"FN={resultats['FN']}  TP={resultats['TP']}")
    print(f"  Coût total      : {resultats['cout_total']:.2f} €")
    print(f"  Coût moyen/indiv: {resultats['cout_moyen']:.4f} €")


# ===========================================================
# 1. Régression logistique classique (seuil 0.5)
# ===========================================================
clf_classique = LogisticRegression()
clf_classique.fit(X_train, Y_train)

proba_classique = clf_classique.predict_proba(X_test)[:, 1]
y_pred_classique = (proba_classique > 0.5).astype(int)

resultats_classique = cout_moyen(Y_test, y_pred_classique)
afficher_resultats("1. Régression logistique classique (seuil=0.5)", resultats_classique)


# ===========================================================
# 2. Régression logistique classique + seuil de Bayes (1/9)
# ===========================================================
# Même modèle que (1), on réutilise les probabilités déjà calculées
y_pred_bayes = (proba_classique > SEUIL_BAYES).astype(int)

resultats_bayes = cout_moyen(Y_test, y_pred_bayes)
afficher_resultats("2. Régression logistique classique + seuil de Bayes (1/9)", resultats_bayes)


# ===========================================================
# 3. Régression logistique pondérée (class_weight) + seuil 0.5
# ===========================================================
clf_pondere = LogisticRegression(class_weight={0: COST_FP, 1: COST_FN})
clf_pondere.fit(X_train, Y_train)

proba_pondere = clf_pondere.predict_proba(X_test)[:, 1]
y_pred_pondere = (proba_pondere > 0.5).astype(int)

resultats_pondere = cout_moyen(Y_test, y_pred_pondere)
afficher_resultats("3. Régression logistique pondérée (class_weight)", resultats_pondere)



# ===========================================================
# 4. Gradient Boosting + seuil de Bayes
# ===========================================================
clf_gb = GradientBoostingClassifier(random_state=42)
clf_gb.fit(X_train, Y_train)
proba_gb = clf_gb.predict_proba(X_test)[:, 1]
y_pred_gb = (proba_gb > SEUIL_BAYES).astype(int)
resultats_gb = cout_moyen(Y_test, y_pred_gb)
afficher_resultats("4. Gradient Boosting + seuil de Bayes", resultats_gb)



````

## J'ai exactement les mêmes résultats et je trouve ça très suspicieux

### On voit clairement qu'un stratégie de pondération est meilleure que la stratégie de base de l'entreprise, ce que je cherche à savoir, c'est pourquoi mes 3 stratégies de pondérations me donnent exactement les mêmes résultats

````python
# Comparer les coefficients et les probabilités, pas seulement les classes prédites
print("Coefs classique :", clf_classique.coef_, clf_classique.intercept_)
print("Coefs pondéré   :", clf_pondere.coef_, clf_pondere.intercept_)

print('\n \n')

#Proba differentes ?
print(np.array_equal(proba_gb, proba_classique))
print(np.array_equal(proba_gb, proba_pondere))

print('\n \n')

diff_proba = np.abs(proba_classique - proba_pondere)
print("Écart max de probabilité :", diff_proba.max())
print("Écart moyen de probabilité :", diff_proba.mean())

# Combien de points changeraient de classe si on perturbait légèrement le seuil ?
proches_frontiere = np.sum((proba_classique > SEUIL_BAYES - 0.02) &
                            (proba_classique < SEUIL_BAYES + 0.02))
print("Points proches de la frontière (zone ambiguë) :", proches_frontiere)
````


### 1-On voit qu'on à bien des coefficients **DIFFERENTS**
### 2-On a aussi des probas différentes
### 3-Changer légérement le seuil de 1/9 ne change rien pour la proba classique : on a pas d'ambiguité autour du seuil

## Le point 3 est la clé


````python
print("Probas identiques ?", np.array_equal(proba_gb, proba_pondere))
print("Prédictions identiques ?", np.array_equal(y_pred_gb, y_pred_pondere))
print("Nombre de désaccords :", (y_pred_gb != y_pred_pondere).sum(), "/", len(y_pred_gb))
````

````python
# Regardez la distribution des probas autour du seuil
import matplotlib.pyplot as plt
plt.hist(proba_pondere, bins=50, alpha=0.5, label='pondéré')
plt.hist(proba_gb, bins=50, alpha=0.5, label='GB')
plt.axvline(SEUIL_BAYES, color='red', linestyle='--', label='seuil Bayes')
plt.legend()
plt.show()

# Combien d'individus sont proches du seuil (zone à risque de désaccord) ?
proche_seuil = np.abs(proba_pondere - SEUIL_BAYES) < 0.05
print("Individus proches du seuil :", proche_seuil.sum())
````

## On comprends que mes modèles, à la fois simple comme regression logistique, ou plus compliqué comme GradientBoosting, sépare tous les deux parfaitement les probabilités pour chaque observations. Il n'y a aucune ambiguité. Cela est surement du au fait que ma variable Contrat permet de bien séparer les individus car on a vu qu'elle impactait fortement le Churn.

Donc si même le modèle simple performe aussi bien qu'un modèle compliqué plus couteux, on va prendre le modèle le plus simple, qui consomme moins en calcul.

# Vérifions nos résultats avec une Cross Validation, pour réduire le risque que nos résultats soient duent au hasard ( on va répter l'opération )

````python
# ===========================================================
# Cross-validation : comparaison des 3 modèles (hors seuil de Bayes)
# manuel, qu'on ne peut pas glisser facilement dans un scorer sklearn
# pour l'instant — on compare donc 1, 3 et 4 directement)
# ===========================================================

#On récrée une fonction qui renvoie un scalaire ( car mona ncienne renvoie un dico, API de sklearn pas content)

def scorer_cout(y_true, y_pred):
    return -cout_moyen(y_true, y_pred)["cout_moyen"]  # négatif car sklearn maximise le score

#On évalue sur la métrique du cout moyen, pas la précision
scorer = make_scorer(scorer_cout)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)         #42 pour noter mes résultats et voir ça fait quoi si je fais des changements

#Nos modèles
modeles = {
    "1. Logistique classique": LogisticRegression(),
    "3. Logistique pondérée": LogisticRegression(class_weight={0: COST_FP, 1: COST_FN}),
    "4. Gradient Boosting": GradientBoostingClassifier(random_state=42),
}

print("=" * 55)
print("CROSS-VALIDATION (5 folds) — coût moyen par modèle")
print("=" * 55)
scores_cv = {}
for nom, modele in modeles.items():
    scores = cross_val_score(modele, X_train, Y_train, cv=cv, scoring=scorer)
    couts = -scores  # on repasse en positif
    scores_cv[nom] = couts.mean()
    print(f"\n{nom}")
    print(f"  Coût moyen CV : {couts.mean():.4f} € (+/- {couts.std():.4f}) ( pour infos : coûts : {couts})")

meilleur = min(scores_cv, key=scores_cv.get)
print(f"\n>>> Meilleur modèle en CV : {meilleur} ({scores_cv[meilleur]:.4f} €)")
````

### Je me rends compte que la cross validation fixe par défault un seuil de eta > 0.5. 

### La bonne idée ici est de créer soit même sa cross validation en essyant d'optimiser un maximum ( surtout au niveau des boucles for, ca peut aller vite avec des modèle de GradientBoosting ou Fôret, on priviligera la vectorisation ) . 

Mais bon, le problème est vraiment très simple, et les résultats de la logistique classique ainsi que la logistique pondérées coîncident avec nos résultats précédents

# 2. Synthèse du projet

L'objectif est de prédire la résiliation (*churn*) des clients télécom à partir de 21 variables. La particularité de ce projet est l'optimisation non pas d'une métrique académique classique (comme l'Accuracy ou le score F1), mais d'un **coût métier réel** défini ainsi :

### 🤖 Modélisation
Trois modèles ont été entraînés et comparés :
1. Régression Logistique classique.
2. Régression Logistique pondérée par les coûts métier.

> **Résultat clé :** Le modèle simple (Régression Logistique) égale les performances du modèle complexe. En effet, la variable `Contract` (type d'engagement) sépare déjà extrêmement bien les deux populations. Il est donc inutile de déployer un modèle coûteux en calcul pour ce problème.

# 4. Présentation vulgarisée pour l'équipe marketing
*Version sans jargon technique, orientée sur l'usage quotidien et l'impact business.*

### 🔮 Qu'est-ce que cet outil fait ?
Il analyse en temps réel les informations d'un client (son ancienneté, le montant de sa facture, son type de contrat, etc.) et calcule la **probabilité qu'il nous quitte dans les prochains mois**. Il agit comme un signal d'alerte précoce.

### 🚀 Pourquoi c'est utile pour vous ?
* **Avant :** Nous traitions tous les clients de la même façon, ce qui dispersait inutilement nos budgets de rétention.
* **Avec l'outil :** Vous pouvez concentrer vos efforts et vos budgets (offres de fidélisation, promotions, appels personnalisés) uniquement sur les clients qui ont un réel risque de partir.
* **Impact financier :** Cela représente une économie estimée à **30 % sur le coût global lié au churn** (chiffre à confirmer sur la version finale).

### 🕹️ Comment l'utiliser concrètement ?
Pour chaque client, l'outil vous transmet deux indicateurs simples :
1. **Un pourcentage de risque :** (ex : 72 % de chance de partir).
2. **Un statut simple :** 🚨 **À RISQUE** ou ✅ **Stable**.

Dans la pratique, ces informations alimentent un tableau de bord où vous pouvez filtrer directement la liste de vos clients "À RISQUE" du mois pour cibler précisément vos campagnes de rétention.

### ⚠️ Ce qu'il faut garder en tête
* **Ce n'est pas une boule de cristal :** C'est une probabilité basée sur des tendances passées. Un client signalé "À RISQUE" ne va pas obligatoirement partir, mais statistiquement, la menace est beaucoup plus élevée que la moyenne. À l'inverse, un client considéré comme "Stable" peut toujours réserver des surprises. L'outil sert à **prioriser**, pas à remplacer votre jugement terrain.
* **Un enseignement commercial fort :** En creusant les données, nous avons constaté que le **type de contrat** est de loin le facteur le plus déterminant. Les clients avec un contrat mensuel (sans engagement à long terme) partent beaucoup plus vite que ceux engagés sur 1 ou 2 ans. Indépendamment du modèle, cela offre déjà une piste de réflexion stratégique pour vos futures campagnes d'offres !
